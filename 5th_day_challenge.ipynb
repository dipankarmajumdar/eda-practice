### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Dataset

```python
df = pd.read_csv("zomato_cleaned.csv")
```

### 1. Bar Chart – Top 5 Restaurant Types

```python
top_rest_types = df['rest_type'].value_counts().head()
plt.figure(figsize=(12,6))
top_rest_types.plot(kind="bar", color="skyblue")
plt.title("Top 5 Resturent Type")
plt.xlabel("Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### 2. Pie Chart – Top 5 Locations

```python
top_locations = df['location'].value_counts().head()
plt.figure(figsize=(6, 6))
top_locations.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 5 locations (Distribution)')
plt.ylabel('')
plt.show()
```

### 3. Histogram – Rating Distribution

```python
plt.figure(figsize=(12,6))
plt.hist(df['rate'], bins=20, color='blue', edgecolor='black')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
